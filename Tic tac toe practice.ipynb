```python
#sahi
# Tic Tac Toe Game
from IPython.display import clear_output
import random

# STEP-1 function to display board
def display_board(board):
    clear_output() # This line is used to clear the output to redraw the board in a cleaner way.
    print(board[7] + '|' + board[8] + '|' + board[9])
    print(board[4] + '|' + board[5] + '|' + board[6])
    print(board[1] + '|' + board[2] + '|' + board[3])

# STEP 2 taking input from player and assigning O X to players
def player_input():
    marker = ''
    while marker not in ['O', 'X']:
        marker = input("Player 1, choose O or X: ").upper()
        if marker not in ['O', 'X']:
            print("Invalid, Try Again!")
    
    player1 = marker
    if player1 == 'O':
        player2 = 'X'
    else:
        player2 = 'O'
        
    return (player1, player2) # It returns a tuple

# STEP 3 function that takes board list obj., marker (O X), position and assigns it to the board
def place_marker(board, marker, position):
    board[position] = marker

# STEP 4 function for all win conditions
def win_check(board, mark):
    # All rows, all columns, and 2 diagonals
    return ((board[1] == board[2] == board[3] == mark) or # row
            (board[4] == board[5] == board[6] == mark) or # row
            (board[7] == board[8] == board[9] == mark) or # row
            (board[7] == board[4] == board[1] == mark) or # column
            (board[8] == board[5] == board[2] == mark) or # column
            (board[9] == board[6] == board[3] == mark) or # column
            (board[7] == board[5] == board[3] == mark) or # diagonal
            (board[9] == board[5] == board[1] == mark))  # diagonal

# STEP 5 randomly choose which player goes first
def choose_first():
    first_chance = random.randint(1, 2)
    if first_chance == 1:
        return "Player 1"
    else:
        return 'Player 2'

# STEP 6 function that returns a boolean indicating if the space on the board is free
def space_check(board, position):
    return board[position] == ' '

# STEP 7 function that checks if the board is full
def full_board_check(board):
    for i in range(1, 10):
        if space_check(board, i):
            return False
    # True means the board is FULL
    return True

# STEP 8 asks player for the next position, then uses step 6 to check if the position is free.
# Then returns the position
def player_choice(board):
    position = 0
    within_range = range(1, 10)
    # Check if both conditions are false i.e., not in range and space is NOT free
    # while loop will run until both conditions are true
    while position not in within_range or not space_check(board, position):
        position = int(input('Enter position (1 to 9): '))
        
        # Invalid message
        if position not in within_range or not space_check(board, position):
            print("Invalid, Try Again!")
        
    return position

# STEP 9 asks user to play again or not
def replay():
    game_on = True
    replay_choice = "Wrong"
    while replay_choice not in ['YES', 'NO']:
        replay_choice = input("Do you want to play again? (Enter YES or NO): ").upper()
        if replay_choice not in ['YES', 'NO']:
            print("Invalid, Try Again!")
    # After selecting replay_choice correctly
    if replay_choice == 'YES':
        game_on = True
    else:
        game_on = False
    return game_on

# MAIN LOGIC
print('Welcome to Tic Tac Toe Game')

while True:
    # Play game
    the_board = [' '] * 10 
    
    # Set everything up (board, who's first, choose marker X, O)
    player1_marker, player2_marker = player_input() # Output of player_input() is getting stored in 2 variables
    print('Player 1 marker:', player1_marker)
    print('Player 2 marker:', player2_marker)
    
    turn = choose_first()
    print(turn + " will go first") # Print who goes first
    play_game = replay()
    if play_game:
        game_on = True
    else:
        game_on = False
    
    # Game play
    while game_on:
        if turn == 'Player 1':
            # Show the board
            display_board(the_board)
            # Choose a position
            position = player_choice(the_board)
            # Place the marker on the position
            place_marker(the_board, player1_marker, position)
            # Check if they won
            if win_check(the_board, player1_marker):
                display_board(the_board)
                print("Player 1 has WON!!!")
                game_on = False
            # Or check for tie
            else:
                if full_board_check(the_board):
                    display_board(the_board)
                    print("TIE GAME!")
                    game_on = False
                else:
                    turn = "Player 2"
        else:
            # Show the board
            display_board(the_board)
            # Choose a position
            position = player_choice(the_board)
            # Place the marker on the position
            place_marker(the_board, player2_marker, position) 
            # Check if they won
            if win_check(the_board, player2_marker):
                display_board(the_board)
                print("Player 2 has WON!!!")
                game_on = False
            # Or check for tie
            else:
                if full_board_check(the_board):
                    display_board(the_board)
                    print("TIE GAME!")
                    game_on = False 
                else:
                    turn = "Player 1"

    if not replay():
        break
```
